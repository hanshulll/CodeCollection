# Doubly LinkedList

```python
class Node:
    def __init__(self):
        self.prev=None
        self.data=None
        self.next=None
        
class doublyLinkedList:
    
    def __init__(self):
        self.head=None
        
    def append(self,data):
        node=Node()
        current=self.head
        if(self.head==None):
            self.head=node
            node.data=data
            current=self.head
            return
        elif(self.head.next==None):
            node.prev=self.head
            self.head.next=node
            node.data=data
            return
        else:
            while(current.next!=None):
                current=current.next
            node.prev=current
            current.next=node
            node.data=data
            return
        
    def display(self):
        print("\n")
        if(self.head==None):
            print("no node")
            return
        else:
            current = self.head
            while(current!=None):
                print(current.data,end="->")
                current=current.next
            return 
    def pop(self):
        current=self.head
        while(current.next.next!=None):
            current=current.next
        print("\n",current.next.data)
        data=current.next.data
        current.next.prev=None
        current.next=None
        return data
    
    def insertAfter(self,data,newdata):
        current=self.head
        while(current.data!=data):
            current=current.next
            if(current.next==None):
                print("Node with data %d not found, unable to append node"%data)
                return
        node=Node()
        node.prev=current
        node.data=newdata
        node.next=current.next
        current.next.prev=node
        current.next=node
        return
            
link=doublyLinkedList()

link.append(10)
link.append(20)
link.append(30)
link.append(40)
link.append(50)
link.append(60)
print("after append:")
link.display()

link.pop()
# link.pop()
print("After pop:")
link.display()

link.insertAfter(40,80)
print("\nAfter insert:")
link.display()
```

```python
class Node:
    def __init__(self,data):
        self.left=None
        self.val=data
        self.right=None

def insert(root, key):
    if root is None:
        return Node(key)
    else:
        if root.val == key:
            return root
        elif root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

def minValueNode(node):
    current = node
    # loop down to find the leftmost leaf
    while(current.left is not None):
        current = current.left
    return current

def deleteNode(root, key):

    if root is None:
        return root
 
    if key < root.val:
        root.left = deleteNode(root.left, key)
 
    elif(key > root.val):
        root.right = deleteNode(root.right, key)
 
    # If key is same as root's key, then this is the node to be deleted
    else:
        # Node with only one child or no child
        if root.left is None:
            temp = root.right
            root = None
            return temp
 
        elif root.right is None:
            temp = root.left
            root = None
            return temp
 
        # Node with two children:
        # Get the inorder successor
        # (smallest in the right subtree)
        temp = minValueNode(root.right)
 
        # Copy the inorder successor's
        # content to this node
        root.val = temp.val
 
        # Delete the inorder successor
        root.right = deleteNode(root.right, temp.val)
 
    return root
 

def inorder(root):
    if root:
        inorder(root.left)
        print(root.val)
        inorder(root.right)
        
root = Node(50)
# root = insert(root, 50)
root = insert(root, 30)
root = insert(root, 20)
root = insert(root, 40)
root = insert(root, 70)
root = insert(root, 60)
root = insert(root, 80)
 
print ("Inorder traversal of the given tree")
inorder(root)
 
print ("\nDelete 20")
root = deleteNode(root, 20)
print ("Inorder traversal of the modified tree")
inorder(root)
 
print ("\nDelete 30")
root = deleteNode(root, 30)
print ("Inorder traversal of the modified tree")
inorder(root)
 
print ("\nDelete 50")
root = deleteNode(root, 50)
print ("Inorder traversal of the modified tree")
inorder(root)
```

```python
def inorder(root,search):
    if root:
        inorder(root.left)
        if(root.val == search):
            return
        inorder(root.right)
    
def preorder(root):
    if root:
        if(root.val == search):
            return
        preorder(root.left)
        preorder(root.right)
        
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        if(root.val == search):
            return
```
